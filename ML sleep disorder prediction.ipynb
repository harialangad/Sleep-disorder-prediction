```python
ob1=LabelEncoder()
ob2=LabelEncoder()
ob3=LabelEncoder()
ob4=LabelEncoder()
ob5=LabelEncoder()

data['Gender']=ob1.fit_transform(data['Gender'])
data['Occupation']=ob2.fit_transform(data['Occupation'])
data['BMI Category']=ob3.fit_transform(data['BMI Category'])
data['Blood Pressure']=ob4.fit_transform(data['Blood Pressure'])
# data['Sleep Disorder']=ob2.fit_transform(data['Sleep Disorder'])
```

```python
x=data.drop(['Sleep Disorder'],axis=1)
y=data['Sleep Disorder']
```

```python
# LOGISTIC REGRESSION
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model1=LogisticRegression()
model1.fit(xtrain,ytrain)
```

```python
pred_y=model1.predict(xtest)
pred_y
```

```python
print(model1.score(xtrain,ytrain))
print(model1.score(xtest,ytest))
```
